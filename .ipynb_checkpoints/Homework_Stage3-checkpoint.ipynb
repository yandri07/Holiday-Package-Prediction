## Model Selection

```python
def plotting_confusion_matrix(model, X_test, y_test, title):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt='g', cmap='Greens', annot_kws={"fontsize":12}, ax=ax)

    # setting title and axis labels
    plt.tick_params(axis='both', which='major', labelsize=12)
    ax.set_xlabel('Predicted labels', fontsize=12)
    ax.set_ylabel('True labels', fontsize=12)
    ax.set_title('Confusion Matrix ' + title, fontsize=14)
    ax.xaxis.set_ticklabels(['Not Taken', 'Taken'])
    ax.yaxis.set_ticklabels(['Not Taken', 'Taken'])
    plt.show()
```

## Model Evaluation

Pada kasus kita bertujuan untuk memprediksi apakah pelanggan akan membeli paket perjalanan atau tidak, maka:
- Metric evaluasi yang akan kita fokuskan adalah **Precision** dimana kita ingin mengurangi jumlah FP (False Positive) artinya model kita ingin menoleransi pelanggan yang diprediksi akan membeli tetapi tenyata aktualnya tidak membeli agar mengurangi biaya pemasaran. Semakin sedikit jumlah FP maka semakin tinggi nilai Precision yang dihasilkan dan jika sebaliknya maka kita kemungkinan hanya mendapatkan sinyal palsu sehingga malah dapat kehilangan pelanggan potensial (nilai Recall tinggi).
- Karena data target pada kasus kita terjadi imbalance class sehingga kita juga akan menggunakan metric **AUC** untuk mengukur apakah model bisa membedakan pelanggan antar kelas. Nilai AUC yang tinggi membuktikan bahwa model yang digunakan sudah cukup baik dalam membedakan kelas negative dan positive.

```python
def eval_classification(model, X_train, y_train, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)
    y_pred_proba_train = model.predict_proba(X_train)
    
    accuracy = round(accuracy_score(y_test, y_pred), 4)
    precision = round(precision_score(y_test, y_pred), 4)
    recall = round(recall_score(y_test, y_pred), 4)
    f1 = round(f1_score(y_test, y_pred), 4)
    
    auc_train = round(roc_auc_score(y_train, y_pred_proba_train[:, 1]), 4)
    auc_test = round(roc_auc_score(y_test, y_pred_proba[:, 1]), 4)
    
    return accuracy, precision, recall, f1, auc_train, auc_test
```

```python
model_name = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'AdaBoost', 'XGBoost']
model_list = [model_lr, model_dt, model_rf, model_adb, model_xgb]

eval_list = []
for name, model in zip(model_name, model_list):
    result = eval_classification(model, X_smote, y_smote, X_test, y_test)
    eval_list.append([name] + list(result))
    
df_model = pd.DataFrame(eval_list, columns=['Model', 'Accuracy', 'Precision', 'Recall',
                                            'F1 Score','AUC (Train)','AUC (Test)'])
df_model
```

Dari hasil evaluasi kelima model tersebut didapatkan bahwa:
- **Model Logistic Regression** mampu memprediksi pelanggan yang membeli paket dengan benar (Precision) sebesar 60.66% dan model sudah cukup baik dalam training dan testing walaupun nilai AUC bisa lebih tinggi lagi.
- **Model Decision Tree** mampu memprediksi pelanggan yang membeli paket dengan benar (Precision) sebesar 68.91% tetapi model masih mengalami overfit terlihat pada nilai AUC pada saat training lebih tinggi sekitar 15% dibandingkan saat testing.
- **Model Random Forest** mampu memprediksi pelanggan yang membeli paket dengan benar (Precision) sebesar 85.83% dan model sudah cukup baik dalam training dan testing dimana nilai AUC sudah cukup tinggi.
- **Model AdaBoost** mampu memprediksi pelanggan yang membeli paket dengan benar (Precision) sebesar 58.87% yang merupakan hasil terendah dari kelima model tersebut dan memiliki nilai AUC yang tidak terlalu tinggi dan sedikit overfit.
- **Model XGBoost** mampu memprediksi pelanggan yang membeli paket dengan benar (Precision) sebesar 86.03% dan model cudah cukup baik dalam training dan testing dimana nilai AUC sudah cukup tinggi.

Maka dapat disimpulkan dari kelima model yang dipilih, **Model XGBoost** merupakan base model terbaik karena memiliki nilai metric yang diharapkan dalam kasus ini (Precision dan AUC) yang paling baik dengan **Precision 0.8603 dan AUC 0.9548** dan metric lainnya yang tidak difokuskan juga paling tinggi diantara model lainnya.

## Thresholding

Dengan melakukan setting threshold yang diinginkan, kita bisa memilih mana yang akan dipilih sebagai kelas 0 (negative) dan kelas 1 (positive) dengan tujuan memaksimalkan pada metric evaluasi yang diambil tanpa mempengaruhi metric yang lain.

```python
# see thresholding to maximize precision and recall
def plot_pr_vs_th(model, X_test, y_test, title=""):
    y_prob = model.predict_proba(X_test)
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    pr, rc, th = precision_recall_curve(y_test, y_prob[:, 1])    
    ax.plot(th, pr[:-1], 'b-', label='Precision')
    ax.plot(th, rc[:-1], 'g-', label='Recall')
    plt.axvline(th[np.argmin(abs(pr-rc))], c='red', ls='dashed')
                           
    ax.set_xlabel('Threshold')
    ax.set_ylabel('Precision and Recall')
    ax.set_title(f'Best Threshold in {title} = {th[np.argmin(abs(pr-rc))]:.3f}')
    ax.legend(loc='lower left')
    plt.show()
```

```python
plot_pr_vs_th(model_lr, X_test, y_test, title='Logistic Regression')
```

```python
plot_pr_vs_th(model_dt, X_test, y_test, title='Decision Tree')
```

```python
plot_pr_vs_th(model_rf, X_test, y_test, title='Random Forest')
```

```python
plot_pr_vs_th(model_adb, X_test, y_test, title='AdaBoost')
```

```python
plot_pr_vs_th(model_xgb, X_test, y_test, title='XGBoost')
```

Dari hasil pengamatan thresholding ini hanya model Random Forest dan XGBoost yang masih dapat memberikan nilai Precision dan Recall yang tinggi sekitar 80% dimana untuk Random Forest dengan threshold=0.4 dan XBoost dengan threshold=0.376

## Hyperparameter Tuning

```python
list_hyperparameters = [
    { # Logistic Regression
        'penalty': ['l2','l1','elasticnet'],
        'C': [float(x) for x in np.logspace(-3, 3, 20)]
    },
    { # Decision Tree
        'max_depth': [int(x) for x in np.linspace(10, 100, 10)],
        'min_samples_split': [int(x) for x in np.linspace(2, 10, 5)],
        'min_samples_leaf': [int(x) for x in np.linspace(1, 10, 5)],
        'max_features': ['auto', 'sqrt']
    }, 
    { # Random Forest
        'n_estimators': [int(x) for x in np.linspace(50, 1000, 50)],
        'max_features': ['auto', 'sqrt', 'log2'],
        'max_depth': [int(x) for x in np.linspace(10, 100, 10)],
        'criterion': ['gini','entropy'],
        'min_samples_split': [int(x) for x in np.linspace(2, 10, 5)],
        'min_samples_leaf': [int(x) for x in np.linspace(1, 10, 5)]
    },
    { # Adaboost
       'n_estimators': [int(x) for x in np.linspace(100, 2000, 1000)],
       'learning_rate': [float(x) for x in np.linspace(0.001, 0.1, 100)],
       'algorithm': ['SAMME', 'SAMME.R']
    },
    { # XGBoost
        'max_depth' : [int(x) for x in np.linspace(10, 100, 10)],
        'min_child_weight' : [int(x) for x in np.linspace(1, 10, 11)],
        'gamma' : [float(x) for x in np.linspace(0, 1, 11)],
        'tree_method' : ['auto', 'exact', 'approx', 'hist'],
        'colsample_bytree' : [float(x) for x in np.linspace(0, 1, 11)],
        'learning_rate' : [float(x) for x in np.linspace(0, 1, 100)],
        'reg_lambda' : [float(x) for x in np.linspace(0, 1, 11)],
        'reg_alpha' : [float(x) for x in np.linspace(0, 1, 11)]
    }
]

def show_best_hyperparameter(model, hyperparameters):
    for key, value in hyperparameters.items() :
        print('Best '+key+':', model.best_estimator_.get_params()[key])
        
def eval_after_hyperparameter_tuning(model, X_train, y_train, X_test, y_test):
    eval_new = eval_classification(model, X_train, y_train, X_test, y_test)
    eval_score = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC (Train)', 'AUC (Test)']
    for score, e in zip(eval_score, eval_new):
        print(score + ': ' + str(e))
```

## Feature Importance

```python
def show_feature_importance(model, title):
    if type(model).__name__ == 'LogisticRegression':
        feat_importances = pd.Series(model.coef_[0], index=X.columns)
    else:
        feat_importances = pd.Series(model.feature_importances_, index=X.columns)
        
    ax = feat_importances.nlargest(17).plot(kind='barh', figsize=(10, 8))
    ax.invert_yaxis()

    plt.xlabel('Score')
    plt.ylabel('Features')
    plt.title('Feature Importance ' + title)
```

```python
# Logistic Regression
show_feature_importance(tuned_lr.best_estimator_, title='Logistic Regression')
```

```python
# Decision Tree
show_feature_importance(tuned_dt.best_estimator_, title='Decision Tree')
```

```python
# Random Forest
show_feature_importance(tuned_rf.best_estimator_, title='Random Forest')
```

```python
# AdaBoost
show_feature_importance(tuned_adb.best_estimator_, title='AdaBoost')
```

```python
# XGBoost
show_feature_importance(tuned_xgb.best_estimator_, title='XGBoost')
```

## Using Random Forest

```python
train_pred1 = tuned_rf2.best_estimator_.predict(X_train2)
test_pred1 = tuned_rf2.best_estimator_.predict(X_test2)
df_train_pred1 = pd.DataFrame({'ProdTaken':y_train2, 'ProdTaken_Pred':train_pred1}, index=X_train2.index.copy())
df_test_pred1 = pd.DataFrame({'ProdTaken':y_test2, 'ProdTaken_Pred':test_pred1}, index=X_test2.index.copy())
```

```python
df_train1 = pd.concat([X_train2, df_train_pred1], axis=1)
df_test1 = pd.concat([X_test2, df_test_pred1], axis=1)
df_final1 = df_train1.append(df_test1).reset_index(drop=True)
df_final1.head()
```

```python
wrong_predict1 = df_final1[df_final1['ProdTaken']!=df_final1['ProdTaken_Pred']]
wrong_predict1['ProdTaken_New'] = wrong_predict1['ProdTaken_Pred'] - wrong_predict1['ProdTaken']
wrong_predict1['ProdTaken_New'] = wrong_predict1['ProdTaken_New'].apply(lambda x: 1 if x<0 else x)

true_predict1 = df_final1[~df_final1.index.isin(wrong_predict1.index.tolist())]
true_predict1['ProdTaken_New'] = true_predict1['ProdTaken_Pred'] + true_predict1['ProdTaken']
true_predict1['ProdTaken_New'] = true_predict1['ProdTaken_New'].apply(lambda x: 1 if x>1 else x)
new_final1 = true_predict1.append(wrong_predict1).sort_index()
```

```python
df_taken1 = pd.DataFrame({'ProdTaken':df_prep['ProdTaken'].unique(),
                         'Count (Before)':df_prep['ProdTaken'].value_counts().values,
                         'Count (After)':new_final1['ProdTaken_New'].value_counts().values,
                         'Percentage (Before)':(df_prep['ProdTaken'].value_counts(normalize=True)*100).values, 
                         'Percentage (After)':(new_final1['ProdTaken_New'].value_counts(normalize=True)*100).values})
df_taken1
```
